Rules governing the distribution of trips in a city
===================================================

In this note I will explain the motivation for each of the rules in the system. The values that should be learned from the data (or picked by the modeler according to some background knowledge) are shown by *???*.


Following the standard conventions, I have partitioned the jobs into *retail* and *non-retail* (find the details about these conventions in another note). This assumptions can be further simplified, if necessary. 

In the following rule, *Total_Density* can be the total number of jobs in an area (or the population of area). The parameter to be learned (or specified) is the ratio of retail jobs for each region type. 
```
retail_employment(area, Density) :-
    business(area, Total_Density),
    Density is ??? * Total_Density. 
    
retail_employment(area, Density) :-
    shopping(area, Total_Density),
    Density is ??? * Total_Density. 
    
non_retail_employment(area, X) :-
    retail_employment(area, Y),
    X is 1-Y.
```

The following rule concerns the trips of non-retail workers which take a taxi to work in the morning:
```
request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    hour(Time, H),
    H >= 7, H =< 10,
    non_retail_employment(To, Density1),
    residential(From, Density2),
    Rate is ??? * Density1 * Density2.
```

I don't know the ratio of retail jobs, and (more importantly) the probability for them to take a taxi to their work. Nevertheless, we have the option of taking them into account separately:
```
retail_start(Time) :-
    hour(Time, H),
    (
        (H >= 7, H =< 10);
        (H >= 13, H =< 15)
    ).

request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    retail_start(Time),
    residential(From, Density1),
    retail(To, Density2),
    Rate is ??? * Density1 * Density2.
```

At the end of working hours, an employee can go back home or go to another area for the purpose of shopping/entertainment. Below are the rules that describe such trips. The first rule is about non-retail employees coming back home:
```
non_retail_end(Time) :-
    hour(Time, H),
    H >=16, H =< 18.
    
request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    non_retail_end(Time),
    non_retail_employment(From, Density1),
    residential(To, Density2),
    Rate is ??? * Density1 * Density2.
    
```

We can assume a different rule for retail workers:
```
retail_end(Time):-
    hour(Time, H),
    H >= 20, H =< 23.
    
request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    retail_end(Time),
    retail_employment(From, Density1),
    residential(To, Density2),
    Rate is ??? * Density1 * Density2.
```

The previous models were about work trips. Since each person has to take these trips regularly, we assume that they are not influenced by seasonal effects (Note that this assumption will be invalid if, for example, people are more likely to take a taxi in a very hot day). 

On the other hand, the trips that are taken in order to receive a service are far more likely to be influenced by such factors. For example, the following rule is about the trips of people who are going out in the weekend:
```
    request(From, To, Time) ~ poisson(Rate) :-
        !(working_day(Time)),
        hour(Time, H),
        H >= 17, H =< 20,
        residential(From, Density),
        recreational(To, Attraction, Time),
        inv_distance(From, To, Inv_Dist),
        peripheral_factors(From, To, Time, F),
        Rate is ??? * Density * Attraction * F.
```
Note that in this rule we have added more parameters that again need to be specified by the modeler or learned from the data. For example the predicate *recreational(To, Attraction, Time)* specifies how attractive a recreational area is at each time. This rule is necessary for example to express the fact that beach is more popular in summer than in winter. The question is that how do we quantify such a difference in popularity?

Same question rises for the predicate *peripheral_factors(From, To, Time, F)*, which is a mixture of other factors:
```
    peripheral_factors(From, To, Time, F) :-
        rain_factor(From, To, F_rain),
        temp_factor(From, To, F_temp),
        F is F_rain * F_temp.
```

The *rain_factor* in turn depends on a parameter. 
```    
    rain_factor(_, _, Time, 1) :-
        not(rains_at(Time)).
    rain_factor(From, To, Time, F_rain) :-
        shopping_area(From),
        residential(To),
        rains_at(Time),
        F_rain is ???.
```
Now if *rains_at(Time)* is a probabilistic fact, we will end up having two Poisson parameters for *request(From,To,Time)* with their corresponding probabilities.
