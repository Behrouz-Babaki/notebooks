Rules governing the distribution of trips in a city
===================================================

In this note I will explain the motivation for each of the rules in the system. The values that should be learned from the data (or picked by the modeler according to some background knowledge) are shown by *???*.


Following the standard conventions, I have partitioned the jobs into *retail* and *non-retail* (find the details about these conventions in another note). This assumptions can be further simplified, if necessary. 

In the following rule, *Total_Density* can be the total number of jobs in an area (or the population of area). The parameter to be learned (or specified) is the ratio of retail jobs for each region type. 
```
retail_employment(area, Density) :-
    business(area, Total_Density),
    Density is ??? * Total_Density. 
    
retail_employment(area, Density) :-
    shopping(area, Total_Density),
    Density is ??? * Total_Density. 
    
non_retail_employment(area, X) :-
    retail_employment(area, Y),
    X is 1-Y.
```

The following rule concerns the trips of non-retail workers which take a taxi to work in the morning:
```
request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    hour(Time, H),
    H >= 7, H =< 10,
    non_retail_employment(To, Density1),
    residential(From, Density2),
    Rate is ??? * Density1 * Density2.
```

I don't know the ratio of retail jobs, and (more importantly) the probability for them to take a taxi to their work. Nevertheless, we have the option of taking them into account separately:
```
retail_start(Time) :-
    hour(Time, H),
    (
        (H >= 7, H =< 10);
        (H >= 13, H =< 15)
    ).

request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    retail_start(Time),
    residential(From, Density1),
    retail(To, Density2),
    Rate is ??? * Density1 * Density2.
```

At the end of working hours, an employee can go back home or go to another area for the purpose of shopping/entertainment. Below are the rules that describe such trips. The first rule is about non-retail employees coming back home:
```
non_retail_end(Time) :-
    hour(Time, H),
    H >=16, H =< 18.
    
request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    non_retail_end(Time),
    non_retail_employment(From, Density1),
    residential(To, Density2),
    Rate is ??? * Density1 * Density2.
    
```

We can assume a different rule for retail workers:
```
retail_end(Time):-
    hour(Time, H),
    H >= 20, H =< 23.
    
request(From, To, Time) ~ poisson(Rate) :-
    working_day(Time),
    retail_end(Time),
    retail_employment(From, Density1),
    residential(To, Density2),
    Rate is ??? * Density1 * Density2.
```

The previous models were about work trips. Since each person has to take these trips regularly, we assume that they are not influenced by seasonal effects (Note that this assumption will be invalid if, for example, people are more likely to take a taxi in a very hot day). 

On the other hand, the trips that are taken in order to receive a service are far more likely to be influenced by such factors. For example, the following rule is about the trips of people who are going out in the weekend:
```
    request(From, To, Time) ~ poisson(Rate) :-
        !(working_day(Time)),
        hour(Time, H),
        H >= 17, H =< 20,
        residential(From, Density),
        recreational(To, Attraction, Time),
        inv_distance(From, To, Inv_Dist),
        peripheral_factors(From, To, Time, F),
        Rate is ??? * Density * Attraction * F.
```
Note that in this rule we have added more parameters that again need to be specified by the modeler or learned from the data. For example the predicate *recreational(To, Attraction, Time)* specifies how attractive a recreational area is at each time. This rule is necessary for example to express the fact that beach is more popular in summer than in winter. The question is that how do we quantify such a difference in popularity?

Same question rises for the predicate *peripheral_factors(From, To, Time, F)*, which is a mixture of other factors:
```
    peripheral_factors(From, To, Time, F) :-
        rain_factor(From, To, F_rain),
        temp_factor(From, To, F_temp),
        F is F_rain * F_temp.
```

The *rain_factor* in turn depends on a parameter. 
```    
    rain_factor(_, _, Time, 1) :-
        not(rains_at(Time)).
    rain_factor(From, To, Time, F_rain) :-
        shopping_area(From),
        residential(To),
        rains_at(Time),
        F_rain is ???.
```
Now if *rains_at(Time)* is a probabilistic fact, we will end up having two Poisson parameters for *request(From,To,Time)* with their corresponding probabilities. 


Weekdays
========

**People going from home to work in the morning. **

<code>
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        residence_area(From, R_density),
        working_area(To, W_density),
        work_start_time(Time, T_degree),
        rain_factor(Time, R_degree),
        Rate is <font color=red>BASE</font> \* R_density \* W_density \* T_degree \* R_degree.
        
</code>

* *R_density* is estimated from population of area (data is available)
* *W_density* is estimated from job density in area (data is available)
* *T_degree* could be interpreted from trip data (could it?)

<code>
    work_start_time(Time, T_degree):-
        Diff is abs(Time-<font color=red>BASETIME</font>),
        T_degree is <font color=red>BASEVAL</font>-<font color=red>SLOPE</font> \* Diff.
</code>

**Instead of degrees, we can use probabilities for R_density, W_density, and T_degree**. But I expect the results to be the same. 

* *rain_factor* is a probabilistic predicate

<code>
    rain_factor(Time, 1) :-
        not(rains_at(Time)).
    rain_factor(Time, <font color=red>DEGREE</font>) :-
        rains_at(Time).
</code>


** Opposite direction: going from work to home in the evening.**
<code>
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        working_area(From, W_density),
        residence_area(To, R_density),
        work_end_time(Time, T_degree),
        rain_factor(Time, R_degree),
        Rate is <font color=red>BASE</font> \* W_density \* R_density \* T_degree \* R_degree.
</code>

* predicates are defined similar to the previous case

** going from work to a service area to receive a service **
<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        working_area(From, W_density),
        service_area(To, S_density),
        work_end_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* W_density \* S_density \* R_degree * D_degree.
</code>

* S_density is defined based on the relative number of single-owner businesses in each area (subjective)
* not sure how to define dist_factor (maybe it should also be based on trip purpose)

** going from home to service areas **
<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        residence_area(From, R_density),
        service_area(To, S_density),
        service_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* R_density \* S_density \* T_degree \* R_degree \* D_degree.
</code>

* *T_degree* should again be interpreted from the trip data. It is likely to span the whole day, and early evening.

** going from one business to another, i.e. job-related trips**
I will assume that this does not depend on rain. But this should be verified. 

<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        working_area(From, W_density1),
        working_area(To, W_density2),
        working_hour(Time, T_degree),
        Rate is <font color=red>BASE</font> \* W_density1 \* W_density2 \* T_degree. 
</code>

**trips between service areas**
for example, by shoppers who go from one mall to another

<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        service_area(From, S_density1),
        service_area(From, S_density2),
        service_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* S_density1 \* S_density2 \* T_degree \* R_degree \* D_degree. 
</code>


** trips from/to recreational areas** I expect these trips to be taken only in afternoon-early evening. 

directly form work:

<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        working_area(From, W_density),
        recreational_area(To, R_attraction),
        work_end_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* W_density \* R_attraction \* T_degree \* R_degree \* D_degree. 
</code>

from home:

<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        residence_area(From, R_density),
        recreational_area(To, R_attraction),
        work_end_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* R_density \* R_attraction \* T_degree \* R_degree \* D_degree. 
</code>

the opposite trip (back home):

<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        recreational_area(From, R_attraction),
        residence_area(To, R_density),
        early_evening(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* W_density \* R_attraction \* T_degree \* R_degree \* D_degree. 
</code>

* not sure how to define *R_attraction* (maybe in terms of area of parks?)
* early_evening is defined similar to *work_start_time*. It can also be defined in a probabilistic way. 

** trips to party areas in the evening **

from home:
<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        residence_area(From, R_density),
        party_area(To, P_attraction),
        party_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* R_density \* P_attraction \* T_degree \* R_degree \* D_degree. 
</code>

from service areas:
<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        service_area(From, S_density),
        party_area(To, P_attraction),
        party_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* S_density \* P_attraction \* T_degree \* R_degree \* D_degree. 
</code>

trips between party areas:
<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        party_area(From, P_attraction1),
        party_area(To, P_attraction2),
        party_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* P_attraction1 \* P_attraction2 \* T_degree \* R_degree \* D_degree. 
</code>


** from party areas back home **

<code> 
    request(From, To, Time) ~ poisson(Rate) :-
        working_day(Time),
        party_area(From, P_attraction),
        residence_area(To, R_density),
        party_time(Time, T_degree),
        rain_factor(Time, R_degree),
        dist_factor(From, To, D_degree),
        Rate is <font color=red>BASE</font> \* R_density \* P_attraction \* T_degree \* R_degree \* D_degree. 
</code>
